## The below code contains graphs of the Google Drive folder **03.Data Analysis/Sb3 Files**

```python
df_1 = "./DRIVE/Sb3 Files/SB3_FILES1_CSV/main.csv" # SB3 files-1
df_2 = "./DRIVE/Sb3 Files/SB3_FILES2_CSV/main.csv" # SB3 files-2
df_3 = "./DRIVE/Sb3 Files/SB3_FILES3_CSV/main.csv" # SB3 files-3
df_4 = "./DRIVE/Sb3 Files/SB3_FILES4_CSV/main.csv" # SB3 files-4
```

### Folder Name: **SB3 files-1**

```python
# CHOOSE DATAFRAME
cur_df = df_1

df = pd.read_csv(cur_df)

# DATA PREPROCESSING
dims = df.iloc[:, 3:12]
dims_names = dims.columns.tolist()
dim_means = dims.mean()
dim_means = dim_means.tolist()
dim_means = [round(n,2) for n in dim_means]

# ADAPTIONS TO GRAPH NAMES
cat = dims_names
val = dim_means

# GRAPH
num_vars = len(cat)
angles = [n * 2 * np.pi / num_vars for n in range(num_vars)]
angles += angles[:1]
val += val[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, val, color='b', linewidth=2)  
ax.set_rlabel_position(0)  

# DESIGN
plt.title(df_1, fontsize=16, color='blue')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(dims)

# ADJUST DESIGN
plt.tight_layout()

# SHOW GRAPH
plt.show()
```

```python
df1 = pd.read_csv(df_1)
df2 = pd.read_csv(df_2)
df3 = pd.read_csv(df_3)
df4 = pd.read_csv(df_4)

df_list = [df1, df2, df3, df4]

# DATA PREPROCESSING
# Filter each df (From Abstraction to Math Operators)
dfs_filt = [df.iloc[:, 3:12] for df in df_list]
# Catch the dimensions names 
dims_names = dfs_filt[0].columns.tolist()

dim_means = [df_filt.mean().tolist() for df_filt in dfs_filt]

# ADAPTIONS TO GRAPH NAMES
cat = dims_names

# GRAPH
num_vars = len(cat)
angles = [n * 2 * np.pi / num_vars for n in range(num_vars)]
angles += angles[:1]
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))

colors = ['b', 'r', 'g', 'm']

for idx, dim_mean in enumerate(dim_means):
    val = dim_mean
    val += val[:1]
    ax.plot(angles, val, color=colors[idx], linewidth=2, label=f'SB3 files-{idx+1}')  

ax.set_rlabel_position(0)  

# DESIGN
plt.title(cur_df, fontsize=16, color='blue')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(cat, fontsize=12, color='black')
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))



# ADJUST DESIGN
plt.tight_layout()

# SHOW GRAPH
plt.show()
```

### The next code includes the **pobability distribution** of the CT Score (Extended) of each folder

### Folder Name: **SB3 files-1**

```python
cur_df = df_1

df = pd.read_csv(cur_df)
ct_score = df['points']

# Gauss bell and save in a dataframe
mean = ct_score.mean()
std_dev = ct_score.std()
min_score = ct_score.min()
max_score = ct_score.max()
x = np.linspace(min_score, max_score, 100)
pdf = 1/(std_dev * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std_dev**2))


plt.figure(figsize=(10, 6))

# HISTOGRAM OF CT Score
plt.hist(ct_score, density=True, color='green',bins=20, alpha=0.75, label='Histogram')

# NORMAL DISTRIBUTION
plt.plot(x, pdf, color='blue', linewidth=2, label='KDE Plot')


# DESIGN
plt.title(f"KDE Plot and Histograms of CT Scores with Probability Distributions of {cur_df}")
plt.xlabel('CT Score (Extended)')
plt.ylabel('Probability Distribution')
plt.legend()


plt.show()
```

## The code bellow is for **Heatmap** of each CT Concept and his frecuency in the SB3 Files

### Folder Name: **SB3 Files 1**

```python
from collections import Counter

cur_df = df_1

df = pd.read_csv(cur_df)
df = df.iloc[:,3:12]

dims_names = df.columns[0:12].tolist()
# CREATION OF FRECUENCY MATRIX
counter_list = [Counter(df[dim]) for dim in dims_names]
max_value = 5
matrix = np.zeros((len(dims_names), 5))
for y, counter in enumerate(counter_list):
    for x in range(max_value):
        matrix[y, x]= counter.get(x, 0)
print(matrix)

# PLOT OF GRAPH
plt.figure(figsize=(8, 6)) 
plt.imshow(matrix, cmap='viridis', interpolation='nearest')
plt.yticks([n for n in range(len(dims_names))], dims_names)
plt.colorbar()

# SHOW NUMERIC VALUE IN EACH HEATMAP CELL
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        val = int(matrix[i, j])
        plt.text(j, i, val, ha='center', va='center', color='white')

# DESIGN
plt.title(f'Heatmap of {cur_df}')
plt.xlabel('CT Score')
plt.ylabel('DrScratch++ CT Concept')
plt.text(1.3, 1.02, 'Frecuency', fontsize=12, ha='center', va='center', rotation=0, transform=plt.gca().transAxes)


plt.show()
```
