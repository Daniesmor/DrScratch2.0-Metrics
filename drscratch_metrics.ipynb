```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
# STYLE FOR PLOTS
sns.set_style('ticks') 
```

## The below code contains graphs of the Google Drive folder **03.Data Analysis/Sb3 Files**

```python
df_1 = "./DRIVE/Sb3 Files/SB3_FILES1_CSV/main.csv" # SB3 files-1
df_2 = "./DRIVE/Sb3 Files/SB3_FILES2_CSV/main.csv" # SB3 files-2
df_3 = "./DRIVE/Sb3 Files/SB3_FILES3_CSV/main.csv" # SB3 files-3
df_4 = "./DRIVE/Sb3 Files/SB3_FILES4_CSV/main.csv" # SB3 files-4
```

### Folder Name: **SB3 files-1**

```python
# CHOOSE DATAFRAME
cur_df = df_1

df = pd.read_csv(cur_df)

# DATA PREPROCESSING
dims = df.iloc[:, 3:12]
dims_names = dims.columns.tolist()
dim_means = dims.mean()
dim_means = dim_means.tolist()
dim_means = [round(n,2) for n in dim_means]

# ADAPTIONS TO GRAPH NAMES
cat = dims_names
val = dim_means

# GRAPH
num_vars = len(cat)
angles = [n * 2 * np.pi / num_vars for n in range(num_vars)]
angles += angles[:1]
val += val[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, val, color='b', linewidth=2)  
ax.set_rlabel_position(0)  

# DESIGN
plt.title(df_1, fontsize=16, color='blue')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(dims)

# ADJUST DESIGN
plt.tight_layout()

# SHOW GRAPH
plt.show()
```

```python
sb1 = pd.read_csv(df_1)
sb2 = pd.read_csv(df_2)
sb3 = pd.read_csv(df_3)
sb4 = pd.read_csv(df_4)

sb_list = [sb1, sb2, sb3, sb4]
```

```python
# CONCAT DATAFRAMES
sb_files = pd.concat(sb_list)
# DROP DUPLICATES AND RESET INDEX
sb_files = sb_files.drop_duplicates()
sb_files = sb_files.reset_index(drop=True)
```

```python
cur_df = "Global Sb3 Files (1,2,3,4)"
df = sb_files

# DATA PREPROCESSING
dims = df.iloc[:, 3:12]
dims_names = dims.columns.tolist()
dim_means = dims.mean()
dim_means = dim_means.tolist()
dim_means = [round(n,2) for n in dim_means]

# ADAPTIONS TO GRAPH NAMES
cat = dims_names
val = dim_means

# GRAPH
num_vars = len(cat)
angles = [n * 2 * np.pi / num_vars for n in range(num_vars)]
angles += angles[:1]
val += val[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, val, color='b', linewidth=2)  
ax.set_rlabel_position(0)  

# DESIGN
plt.title(cur_df, fontsize=16, color='blue')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(dims)

# ADJUST DESIGN
plt.tight_layout()

# SHOW GRAPH
plt.show()
```

```python
df1 = pd.read_csv(df_1)
df2 = pd.read_csv(df_2)
df3 = pd.read_csv(df_3)
df4 = pd.read_csv(df_4)

df_list = [df1, df2, df3, df4]

# DATA PREPROCESSING
# Filter each df (From Abstraction to Math Operators)
dfs_filt = [df.iloc[:, 3:12] for df in df_list]
# Catch the dimensions names 
dims_names = dfs_filt[0].columns.tolist()

dim_means = [df_filt.mean().tolist() for df_filt in dfs_filt]

# ADAPTIONS TO GRAPH NAMES
cat = dims_names

# GRAPH
num_vars = len(cat)
angles = [n * 2 * np.pi / num_vars for n in range(num_vars)]
angles += angles[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

colors = ['b', 'r', 'g', 'm']

for idx, dim_mean in enumerate(dim_means):
    val = dim_mean
    val += val[:1]
    ax.plot(angles, val, color=colors[idx], linewidth=2)  

ax.set_rlabel_position(0)  

# DESIGN
plt.title(cur_df, fontsize=16, color='blue')
ax.set_xticks(angles[:-1])
ax.set_xticklabels(dims)

# ADJUST DESIGN
plt.tight_layout()

# SHOW GRAPH
plt.show()
```
