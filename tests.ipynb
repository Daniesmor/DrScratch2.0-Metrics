```python
# Datos
categorias = ['A', 'B', 'C', 'D']
valores = [20, 35, 30, 25]

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))  # Tamaño opcional de la figura
plt.bar(categorias, valores, color='g')

# Personalizar el gráfico
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.grid(axis='y')  # Mostrar cuadrícula en el eje Y
plt.tight_layout()  # Ajustar diseño

# Mostrar el gráfico
plt.show()
```

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [4, 3, 2, 5, 4]

num_vars = len(categorias)
angulos = [n * 2 * np.pi / num_vars for n in range(num_vars)]
angulos += angulos[:1]
valores += valores[:1]
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angulos, valores, color='b', linewidth=2)  
ax.set_rlabel_position(0)  

ax.set_xticks(angulos[:-1])
ax.set_xticklabels(categorias)

# Ajustar el diseño
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
